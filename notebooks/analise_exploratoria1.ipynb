Calcule a média de avaliações (ratings) por categoria para identificar as categorias mais bem avaliadas.

```python
# Substituir ou remover valores não numéricos na coluna de avaliação
df_selecionado['rating'] = pd.to_numeric(df_selecionado['rating'], errors='coerce')

# Remover linhas com valores não numéricos (NaN após a conversão)
df_selecionado = df_selecionado.dropna(subset=['rating'])
```

```python
# Converter a coluna de avaliações para o tipo float
df_selecionado.loc[:,'rating'] = df_selecionado['rating'].astype(float)

# Calcular a média de avaliações por categoria
media_por_categoria = df_selecionado.groupby('category')['rating'].mean().reset_index()

# Ordenar por avaliação média em ordem decrescente
media_por_categoria = media_por_categoria.sort_values(by='rating', ascending=False)

print(media_por_categoria)
```
